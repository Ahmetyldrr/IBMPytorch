```python
import pandas as pd
import plotly.express as px

# df veri Ã§erÃ§evenizin mevcut olduÄŸunu varsayalÄ±m
# df = pd.read_csv("dosyaniz.csv")

# 1. HiÃ§ NaN olmayan sÃ¼tunlarÄ± atla: 
#    df.isnull().any() her sÃ¼tun iÃ§in True/False dÃ¶ndÃ¼rÃ¼r. True olan sÃ¼tunlarda en az bir NaN vardÄ±r.
cols_with_nan = df.columns[df.isnull().any()]

# Bu sÃ¼tunlarÄ± iÃ§eren bir alt veri seti oluÅŸturalÄ±m
df_with_nan = df[cols_with_nan]

# 2. NaN maskesi oluÅŸtur ve 0/1'e dÃ¶nÃ¼ÅŸtÃ¼r
nan_mask = df_with_nan.isnull()
nan_mask_int = nan_mask.astype(int)

# 3. GrafiÄŸi oluÅŸtur: 
fig = px.imshow(
    nan_mask_int,
    color_continuous_scale=["yellow", "red"],  # 0: Veri mevcut, 1: Veri eksik
    aspect="auto",
    height=900,  # Dikey boyut: Ã‡ok veriniz varsa artÄ±rÄ±n
    width=800
)

fig.update_layout(
    title="NaN Ä°Ã§eren SÃ¼tunlardaki Eksik DeÄŸerler",
    xaxis_title="SÃ¼tunlar",
    yaxis_title="SatÄ±rlar"
)

fig.show()
```

```python
y = df['Target']
X = df.drop('Target', axis=1)
X
```

```python
from sklearn.preprocessing import OrdinalEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer

ordinal_cols = ['Age in 5-year groups',
                'Highest educational level',
                 'Wealth index combined']


nominal_cols = ['Type of place of residence',
                'Have mosquito bed net for sleeping (from household questionnaire)',
                "Smokes cigarettes",
                "Current marital status",
                "Currently residing with husband/partner",
                "Had fever in last two weeks",
               "Taking iron pills, sprinkles or syrup"]

ordinal_encoder = OrdinalEncoder()
one_hot_encoder = OneHotEncoder(drop='first', sparse=False)

preprocessor = ColumnTransformer(
    transformers=[
        ('ordinal', ordinal_encoder, ordinal_cols),
        ('onehot', one_hot_encoder, nominal_cols)
    ],
    remainder='passthrough'
)

# Veri Ã§evirimini uygula
X_encoded = preprocessor.fit_transform(X)
X_encoded
```

```python
import numpy as np
import pandas as pd
from mord import LogisticIT
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y_encoded, test_size=0.2, random_state=42)

model = LogisticIT()
model.fit(X_train, y_train)  # EÄŸer burada hala hata alÄ±yorsanÄ±z, daÄŸÄ±lÄ±mÄ± tekrar ayarlayÄ±n.

y_pred = model.predict(X_test)

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))

print("\nClassification Report:")
print(classification_report(y_test, y_pred, digits=3))
```
