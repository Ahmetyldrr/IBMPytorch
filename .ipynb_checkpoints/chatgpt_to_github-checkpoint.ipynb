```python
import json
import requests
from colorama import Fore, Style
import json
import pandas as pd
import ssl
import http.client

class ConversationProcessor:
    def __init__(self, conversation_id):
        """
        Initializes the ConversationProcessor with the given conversation ID and authorization token.

        :param conversation_id: The ID of the conversation to process.
        :param auth_token: Bearer token for authorization.
        :param host: The API host. Default is 'chatgpt.com'.
        """
        self.conversation_id = conversation_id
        self.headers ={
  
    'accept': '*/*',
    'accept-language': 'tr-TR,tr;q=0.9,en-US;q=0.8,en;q=0.7',
    'authorization': 'Bearer eyJhbGciOiJSUzI1NiIsInR5cCI6IkpXVCIsImtpZCI6Ik1UaEVOVUpHTkVNMVFURTRNMEZCTWpkQ05UZzVNRFUxUlRVd1FVSkRNRU13UmtGRVFrRXpSZyJ9.eyJzZXNzaW9uX2lkIjoiaVdEeWJYbUl3LTJlbUtNTGdhWDdidDc1WWRmak9EanMiLCJodHRwczovL2FwaS5vcGVuYWkuY29tL3Byb2ZpbGUiOnsiZW1haWwiOiJ5aWxkaXJpcmFobWV0MjFAZ21haWwuY29tIiwiZW1haWxfdmVyaWZpZWQiOnRydWV9LCJodHRwczovL2FwaS5vcGVuYWkuY29tL2F1dGgiOnsicG9pZCI6Im9yZy1JYW9wZFFtZ1UySnRPYjBheE9wVkFqUTUiLCJ1c2VyX2lkIjoidXNlci1tNHFoeWFnVE9xMXk5dGlpU3BnWVM4REkifSwiaXNzIjoiaHR0cHM6Ly9hdXRoMC5vcGVuYWkuY29tLyIsInN1YiI6Imdvb2dsZS1vYXV0aDJ8MTA2NzQwNzQwNjIyNzA3NTEzNDY2IiwiYXVkIjpbImh0dHBzOi8vYXBpLm9wZW5haS5jb20vdjEiLCJodHRwczovL29wZW5haS5vcGVuYWkuYXV0aDBhcHAuY29tL3VzZXJpbmZvIl0sImlhdCI6MTczMzExNzQ4MywiZXhwIjoxNzMzOTgxNDgzLCJzY29wZSI6Im9wZW5pZCBlbWFpbCBwcm9maWxlIG1vZGVsLnJlYWQgbW9kZWwucmVxdWVzdCBvcmdhbml6YXRpb24ucmVhZCBvcmdhbml6YXRpb24ud3JpdGUgb2ZmbGluZV9hY2Nlc3MiLCJhenAiOiJUZEpJY2JlMTZXb1RIdE45NW55eXdoNUU0eU9vNkl0RyJ9.TKU-tXtQt_etHtbYBasomQmijnN4N0NVvoMv7LgPPM4ZA6VfQyJFHiBFfxZb-AvoKXLcfJi4z4BU_vs1yim7xxwY85wptRL0mbhYXDOwN_FHrCHyoVA4139kQVX7kqSEwJ9qaGrOs44u0OlYPs8PvL0Mr-CeSmqjTkBYGDlRG-wWfipHAD_e1fO--18xx60gXI5kyjJ0UPRd732mt8V3KEaXd3rCp2AlClEBcIowRNJ8Y5NR5SCrMRnm5VQ6YwVRizf7MrXwPNYWeRWvxdq7MIFj8YLS_wAHHKD6IwDcFL3ctD0m3C2uLLD48KIh8Q3G5p7m4ly17QfeDbhLrlRk_A',
    'cache-control': 'no-cache',
    'cookie': 'oai-did=b17603b7-ff55-46d8-911d-629c5dc08d05; oai-hlib=true; oai-nav-state=1; oai-sidebar=closed; _puid=user-m4qhyagTOq1y9tiiSpgYS8DI:1733136578-%2Bv4u48JXB0KRMsohAOqSr825vtsqOIZV6FYeqwqgQHM%3D; __Host-next-auth.csrf-token=d4e4b176b10bec49279176dab084c284de16602916ec3afef03b6e5c80078e12%7Cb18530379119f8bab18e31085b6a07f3c2cf75d818506fc87859cbda0831b4c2; __Secure-next-auth.callback-url=https%3A%2F%2Fchat-identity-edge-onramp.unified-5.api.openai.com; _cfuvid=MIGUSWANV4oD9.DkDRs.UdxFemLzFeP9JX62ILNksYM-1733222112686-0.0.1.1-604800000; __cflb=0H28vzvP5FJafnkHxiseZXjqM4UTrUP2J6K9KjZHjq7; __cf_bm=0S5wpsNVTVlaVGmE5_4v.cq0zRl1pGVp7k8fvCzRo_o-1733224422-1.0.1.1-MW.4I1ocovP2unRX7ZlEl0BGZ.FpAsk4RLxuJkhEr8ppMfQ9gA0k.I4oEuN2UnB9f.Azstspl.xZxBzSuaEI4w; _uasid="Z0FBQUFBQm5UdWsyM3R0ZjRSbWdqVG44Sy1zVENtbGxtTkpndDdmdEt5SkRGcExSRGQzTy0xSjBZdjIxMGFZcWRRUlhDTW5TMDVGR3UwQkh5dXQ4T0RKdTVLWDRZU1QtNnhsM3JzZkJ3UFdpcm8tU2YzdVAwTWRnYjY4ZnNHWU5aVFl6SjZLNEMtdG0tdXE0YzZyeVlPZ0l4bnN6UkUzMWY2MWFrQ29IeVdRMC01RzQyWTZMbDhqSkJvUEZVZk0zYlJXcWJadGVyNFg0Z09tN1U3VWVFWjBEUWFITmVNUEdXWDgxckxGRzFIZWgxSFViUlJiSFVSbEZ4MkdmQjl5Rl81b0pSMXR5ZndydG1EejRQaTR1ZERlSFJweHdKcWRJOUFCOHgzWEdTMVlvRDAxQ3lybzZidXY5YmJVcE5vZ2ZrdkZXMlU4WDhFZmtSdWdaa21GM3drRXdTZTUyYl9XZjRRPT0="; _umsid="Z0FBQUFBQm5UdWsyWTFZNzFwclg5Y3FXcTR4TDgzNC14WTh6U09nS0VKVTN0U0pVV1MtVVhWYlhOLVRRa2twQ3o5RVc5RnVnZkozSXNGMm1Va1VwYWN4UzhLUlA4WmxISG5vTWowTmstQVR5ZVN0c3VEMUhKRkJBcmZsQ1YtckhmcmxjX3ViMXRveVZmUTJ5Q0t6YlJoajlFNVVvSUFsY0dKNng3elJ1Qi1hRkhySFJzanlEdHc0UE9jV3RsRHF0aHZ6cVkwZzNDdzJUNVhnLUhNMEdhWVRYVjI0MjF1TE0yQTVPRklkV1BZM1liVmFZLTJWU0dfYz0="; oai-sc=0gAAAAABnTuk24J4Dc8M9YB8mE68yG1O7XX3pyvIq2lZuV7bCOfJ44iDMhSEPUSbxhKmjgT1z0il2dvzzbIdVECI5Y__AWP4m_wo0q0FJ2KTTyWzWMqTri0bNGfTvM7SVAOuFErJetqK1p5QRvuuJJjOMfZOoYhO1AaaG8viHC8_pSO2xG9LbXN4Q3Cc-4WJ5jBUCrdd1_Wa-yM0IkwvPGtL6zuRmlfkYCMq8BoJqpd0QiCr-0xxsYD4; __Secure-next-auth.session-token=eyJhbGciOiJkaXIiLCJlbmMiOiJBMjU2R0NNIn0..eb96c0CnKEanTCkm.Ijj66IRqdby5UuTb3m99HYp-2VMNSt7uCwO8oba3x7goFu8OzUTAYOLrVRiLcymhBbTdfAmQECKXbodQr7t5H0qLCXtRMb6_4jYAKXjdd3iAY0q21RipQ2IoarlK8TChJWwxe29sbYoW_wSSOSoXrUJIqM84HmGWwL_60CZeb5F84KLk3lbHOx0KQTi-z9n8EiubVKhVXaOqn3qxY4zsDtx1q1VLWcZDoK0Rua_WjYm2QV5u1xk-6l-RvRDAQZ1yoiIhGwKyDN8FH7mpJ8-xy-MfGUBBsvtaTFLT7UD-BemjdtVYSznUSZNTSapH1XQ38BZfQiIHjMOTtBlQHbXihhf5zilSvEhtYN8Prt68B2W2BxKK3me0jcXstgLf9iN0Yiyrox-nUwtO8CRWmvVQnr-hMIbCHtMGoJLXJtdWeQSM1YN_3rrcsGPL6AOh1avdUG1xr28m_OIrYHoeVl9Ivm0ZoVZOmtn2pWAJdV42qdfzTRiavBEPgiBW4iY_fdNxvQ8ZjZV7d7res5PSDL-ZBIFU0IZ_P0lnI7oAWJmGv4fPrGrtzVlW2ttkjeZx2hctZ5wcc5Ysv50pWU5_fvycVyJkxhxYgHxLeHD5QLz911Wfe3RZT0feoSZ5M9xpFlnVGYM7BawNk2i1tSSdxt4RBSbR5-npyoqr6fvDqap5LVh9yF1Y7IQnt4W1kT_XlWr_s6pTzq8o2cnar97kDel1-2Pp4zIKESuHRDRRVCiFscaDyM-o3yZZVBmpRn5AtfPqSPIgUO5CPHIsrlzbxrKaRFSnLsM9xNwMPI2cm9dgcXB9ZLJnZmoOM7nQNPfTlureaLMFwSt0Kl9n1D1HXx4yV3D_iIJusEretOvMeUaQwzNfPcZFXTx62XbGR8c_wv__P6LWe_8sv694iXooV0D8ZJABHs1mP0SOdkl_AT7-242xrbS8jxRf-p9gBNOVJnIqVzkdQZoAdyewBFZf8QU63p7ssR0AJVKsCHcmSGPvWaq044Gwt8hIbX9ZuS9kq3b7viMrRYTEbGX2BOmNeULWFuLth1NzdKg6xiicabKZ7fom5XOep1cUr04JGaPEOpqTsr7wVDgPg-Qy9_KZHKiSRm3nOE_8dRlGXpWRE1OVK01ULtllc85Gule_jCi1WqB_p7hWyxFvMb7WWTu6Bvtwv3xjJc-x52lIA8r0NDznfcaxRLN87EvTJViSKzTvWVmDfxlLHOOOINrzjS8QAKLURyOP1BK1F_Lbeyv-seQbzWY83SKDDzXKvEDUo4CBwQp_X9qatQt_a-PkKSD6oE5acKlhnFFRSV5-9cBSZv1I1Ubssb44VKFH2Lx8uwAsNMCV7_zJZInNIGzYjJBEPoYZHKTEKPzXcRRpy3OqlhD1LmKLndQfYXGfWmBLe62MZEZlJZMIv9l3j2u4rnqSSaYlUm4ttHIvO1U2oYjM2avvRvd7QtnFwE9ZuXY5XXLqYADaDNHHnaGR8g8OpbtkA3Pnn6tGmf9hjaIuHYeCB50hmcA4WliCJ_bPxgZKq1JvwNsDt-2-mTClxRgmH5ORb5g3gx-RpfegvyngDgHoaCxv3Ei9g2DCYoRsIyNpgHXSUUNjsZvselBsbvmLL843_9Sx5SlDFpbbF-Eyx_4K0F9sHGAcNaAl5lvbITmwMCFjvWdnMmUNJLnWgUcEo0oCpslxh5Q8Q6D6hF_pT2tAQPLs2gKlovEBRvJ3wfJEcYtC9fHwLstI8E21_VB5CBn60_TNptLR9QxUXI0lzti56XMwZgV2qPi3RXhLkXFtGvxM_DXZWVXexF8cNpeQ74AZ_BaPyYVCeD9WFYgKi2TIg6p6IDYmmt8bkUUgyYq3ao-3GSQkEhwbzK1CTcE3rDvhfhJ-rB2Ut3N6_0lGmogdS0FBPyDKvvzATkftpqlYDBoXmLc2KkDWzqv4eSYe5QU_c98RQ_SEpC-v6KUhv73vas4XtAbNlx-LFnKLklvevAeqK1qdyfx3rYx22_7uZmQfaTFjz_qwCTVPwjxgtpd-oxzfGIrkvpwKqaPtiD4Ao3RZgEKDazyKJBx-oeqpLeuTmbVPKDYhRRyk2VcVP2BUrzOBNBWLUCWogWZikZVJ7kAZTb107f9775Ku2Gunjz6lMZ4xge70lAIVXxzhLZ3fTc0zVx5pBSfRjDVxEdZRmVIrADL2aGO877cpZGSfK05F6R8s9niW-_ROo1UI2NBaIfMRG9HEkDekwhkrg1QRBcXpGiTqgeYyioF-TPwwa_bHAuoD-546CC3Qd2LmcXz_H7NmoptJsLYaF3u4s9fppI2BwNHwrDXkfHlYVoj8pyKcSj_zyhpmxNu7voLlY4F2vrb-sq8e0h2R4C79YVZeypy1ORrU6uBQX8MzzVS7eB3O6C1-07xdFmZYKCZcIeUWYWD3wdY-_MePQIe___mM-xAMVKMAlqYE1h33PHBFuDpr_b5Gm-ztR7aNJck0-75ubtwVT68FLwGDJ9NXbJYmy31ue8XAhAirtrIWoLITrVFkNSEdUu3wZqrEzdfHlqTAik4NVNdvO-K_FYUV0otv6lZ9qJSU4atAitoIDJ8MI-zlzuVzuUyxVN53574LvedXxkC15-H8Awbb-VLpSoPXI80dsMqATQju1WIoyX5ARWVv4HM3zPFY-P4Elg3cJdA8bdAo-iCHALRHbr6do2KfM0b-Pg4x96ewC0MnCNulhf_lWkM89OYwHh3px7p69i2lz6izIiReiTdaC-mXL8kgRixHOydiRExknB8pZT0X0a7iGraRynC7TfXxByjM5_3Uj_E5-03karwnDSx7bfmXDLxM88gTsY85aQ-tsm_PniiyQejE5DVSrp7dKX0Qy2sCBYR7OepSZkCKxnz2FheT4s6fgj6z0M_kDZ9Boh-oWd6VeAztsDt8abU1bQXysiSZI-khPv6IMyiIYC1pmPznwy-1eV0h.NM3dxbkVs1rlzzmyVkF8og; cf_clearance=V0ApSeS9xKYmqLgWYbslHHcInYzFviUIV.qSAnTsrZg-1733225095-1.2.1.1-Ssl0V7tIXCO4qyx2EpQGaJQqFLn6FN7EvMt3XdYRC9MfmniQsl40lqh6j8pUtdZOsyxxxO3GgPwQphpQmAHM1LvaVV_BCe2uJm88c5.IoHUmWvWl.2kJdd8LR4XVY7nTokskgGa305c5l_9sFkjPMxgrDcxzOMHiAagL82VXWXHEBVdzI7QfX4IC1C4DWqktXwJ6Z3oPOy.95jrfsysMYdyZoS8grehhZQ0G071pYZ2cPu5469aE3jR3f8WlOj7gZmCQHncViIBh8oHk0ogSTlH9TMTv4KfjgHTXvDxYSX.t8dj8ujz.WpMsbdj6.Ygt9wDj_hy2Shzt_qe7tzezDu4td6behXbxt0gNERHJRvsJxmY2rgpGvvlHbH8y6fuRMG2MZrAXO_7D7wr1c_WPHA; _dd_s=rum=0&expire=1733225995381&logs=1&id=55cfa0b8-b39b-4276-b7b3-0112058eb415&created=1733222113725',
    'oai-device-id': 'b17603b7-ff55-46d8-911d-629c5dc08d05',
    'oai-language': 'tr-TR',
    'pragma': 'no-cache',
    'priority': 'u=1, i',
    'referer': 'https://chatgpt.com/c/674ee935-c3a0-800e-b92d-c584249c82a3',
    'sec-ch-ua': '"Google Chrome";v="129", "Not=A?Brand";v="8", "Chromium";v="129"',
    'sec-ch-ua-mobile': '?0',
    'sec-ch-ua-platform': '"Windows"',
    'sec-fetch-dest': 'empty',
    'sec-fetch-mode': 'cors',
    'sec-fetch-site': 'same-origin',
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/129.0.0.0 Safari/537.36',

}

            
                    

 
        
    def fetch_conversation_data(self):


        try:
            context = ssl._create_unverified_context()
            conn = http.client.HTTPSConnection('chatgpt.com',context=context)
            conn.request(
                'GET',
                '/backend-api/conversation/'+str(self.conversation_id),
                headers=self.headers
            )
            response = conn.getresponse()
            
            data = response.read()
            data = json.loads(data)
           
            return data
            
        except requests.exceptions.HTTPError as http_err:
            print(f'HTTP error occurred: {http_err}')
            
        except Exception as err:
            print(f'Other error occurred: {err}')
            return None

    def traverse_messages(self, mapping, current_id, messages_x):
        """
        Recursive function to traverse the message tree and label messages.

        :param mapping: The mapping dictionary from the conversation data.
        :param current_id: The current message ID being traversed.
        :param messages_x: The list to append labeled messages.
        """
        message_info = mapping.get(current_id, {})
        message = message_info.get('message')

        if message and message.get('content'):
            try:
                author_role = message['author']['role']
                content = message['content']['parts'][0].strip()
                if content:  # Boş içerikleri atla
                    if author_role == 'user':
                        # "Soru:" etiketini daha kalın hale getirmek için üç yıldız kullandık
                        labeled_message = f"***Soru:*** {content}"
                    elif author_role == 'assistant':
                        # "Cevap:" etiketini kalın yapmak için çift yıldız kullandık
                        labeled_message = f"**Cevap:** {content}"
                    else:
                        labeled_message = f"{author_role.capitalize()}: {content}"
                    messages_x.append(labeled_message)
            except:
                pass

        
        children = message_info.get('children', [])
        for child_id in children:
            self.traverse_messages(mapping, child_id, messages_x)

    def get_labeled_messages(self):
        """
        Retrieves and processes the conversation data to return labeled messages.

        :return: A string containing all labeled messages separated by two newlines.
        """
        data = self.fetch_conversation_data()
        if not data:
            return "Veri alınamadı."

        mapping = data.get('mapping', {})
        if not mapping:
            return "Mesajlar bulunamadı."

        # Mesajları etiketlemek için boş bir liste oluştur
        messages_x = []

        # Kök düğümü bulmak (parent = None)
        root_ids = [msg_id for msg_id, msg in mapping.items() if msg.get('parent') is None]

        # Her kök düğüm için traverse işlemini başlat
        for root_id in root_ids:
            self.traverse_messages(mapping, root_id, messages_x)

        # Etiketlenmiş mesajları yeni satırlarla birleştir
        message = "\n\n".join(messages_x)
        return message


conversation_id = "674ee935-c3a0-800e-b92d-c584249c82a3"
processor = ConversationProcessor(conversation_id)
labeled_messages = processor.get_labeled_messages()

```
