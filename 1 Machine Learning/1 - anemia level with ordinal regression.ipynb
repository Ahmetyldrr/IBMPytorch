- **Age in 5-year groups**: 5 yıllık yaş aralıkları (Örnek: 15-19, 20-24, ...)
- **Type of place of residence**: Katılımcının yaşadığı yerin türü (kentsel/kırsal)
- **Highest educational level**: Sahip olunan en yüksek eğitim seviyesi (örnek: ilkokul, lise, üniversite)
- **Wealth index combined**: Varlık endeksi (hanehalkının sosyoekonomik durumuna göre gruplar)
- **Births in last five years**: Son 5 yılda gerçekleşen doğum sayısı
- **Age of respondent at 1st birth**: İlk doğumda anne yaşı
- **Hemoglobin level adjusted for altitude and smoking (g/dl - 1 decimal)**: Rakım ve sigara kullanımına göre ayarlı hemoglobin düzeyi
- **Anemia level**: Anemi derecesi (örneğin: normal, hafif, orta, şiddetli)
- **Have mosquito bed net for sleeping**: Hanehalkının sivrisinek cibinliği sahipliği (Evet/Hayır)
- **Smokes cigarettes**: Sigara içme durumu (Evet/Hayır)
- **Current marital status**: Güncel medeni durum
- **Currently residing with husband/partner**: Eşi/partneri ile birlikte yaşama durumu (Evet/Hayır)
- **When child put to breast**: Çocuk göğüse ne zaman konuldu (doğumdan hemen sonra, birkaç saat sonra vb.)
- **Had fever in last two weeks**: Son iki haftada ateş geçirdi mi? (Evet/Hayır)
- **Hemoglobin level adjusted for altitude (g/dl - 1 decimal)**: Yüksekliğe göre ayarlı hemoglobin düzeyi
- **Anemia level.1**: Anemi derecesinin alternatif bir ölçümü
- **Taking iron pills, sprinkles or syrup**: Demir desteği (hap, damla, şurup) alımı
